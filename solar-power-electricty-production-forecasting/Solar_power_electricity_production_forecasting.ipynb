#### Taking necessary columns in a new dataframe

```python
y=merge_df[['Time','production MW']]
y
```

#### Setting time column as index

```python
y=y.set_index(['Time'])
y
```

#### calculating statistics

```python
y.info()
```

```python
#Determine rolling statistics
rolmean = y.rolling(window=15).mean() #window size 12 denotes 12 months, giving rolling mean at yearly level
rolstd = y.rolling(window=15).std()
print(rolmean,rolstd)
```

```python
## plot graph
plt.xlabel('Date')
plt.ylabel('Nuymber of production MW')
plt.plot(y['production MW'].index[200:296])
```

#### plotting statistics with trend and seasonality

```python
#Plot rolling statistics
plt.figure(figsize=(15, 4))
orig = plt.plot(y, color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.xticks(rotation=30, ha='right')
plt.show()
```

#### Dickey Fuller Test

```python
from statsmodels.tsa.stattools import adfuller

#Perform Augmented Dickey–Fuller test:
print('Results of Dickey Fuller Test:')
dftest = adfuller(y['production MW'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

## Using fbprophet 

```python
df = merge_df[['Time','production MW']]

df.rename(columns={"production MW":"y",'Time':"ds"},inplace=True)
df
```

```python
Train_df=df.iloc[0:4897,:]  #train dataframe
Test_df =df.iloc[4897:,:]   #test dataframe
```

```python
Test_df
```

### Modelling 

```python
from prophet import Prophet
```

```python
model=Prophet(seasonality_mode='additive',uncertainty_samples=100,daily_seasonality=False,changepoint_prior_scale=0.5).add_seasonality(name='15 min', period=1, fourier_order=40)
model.fit(Train_df)
```

```python
model.params
```

```python
Test_df.reset_index(inplace=True)
Test_df.drop(["index"],axis=1,inplace=True)
```

```python
forecast=model.predict(Test_df)
# forecast=forecast[["ds","yhat"]]
forecast
```

```python
# Plot the components of the model
fig = model.plot_components(forecast)
```

```python
final_df=pd.concat((forecast["yhat"],Test_df),axis=1)
final_df
```

```python
pd.to_datetime(final_df['ds'])
```

```python
def replace(x):
    if x<0:
        return 0
    else:
        return x
```

```python
final_df['yhat']=final_df['yhat'].apply(replace)
final_df
```

```python
final_df
```

```python
from prophet.diagnostics import cross_validation
df_cv = cross_validation(model, initial='30 days', period='50 days', horizon = '10 days')
```

```python
df_cv
```

```python
# Importing performance_metrics from diagnostics

from prophet.diagnostics import performance_metrics

df_p = performance_metrics(df_cv)
df_p
```

```python
# Evaluating our model
from sklearn.metrics import mean_squared_error
import math

mse = mean_squared_error(final_df['y'],final_df['yhat'])
print(math.sqrt(mse))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(final_df['y'],final_df['yhat'])
mape
```

```python
plt.figure(figsize=(20,6))
plt.plot(final_df["ds"],final_df["y"],color="red",label="actual")
plt.plot(final_df["ds"],final_df["yhat"],color="blue",label="forecast")
plt.xticks(rotation=30, ha='right')
plt.legend()
```
